```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# We are training tree based models ,Scaling operations are not required.

# Test Size = 20% -- 20% of data if going to be in the test set , 80% going to be in the training set
# X train , X test , Y train , Y test
# Train Test Split -  Split our data as training and testing sets -- Training Set - Training ML models , also evaluvated the models performance with the test set so we are going to have 4 features.

```

```python
from sklearn.model_selection import GridSearchCV

# Grid Search allows you to try different Paramters for ML models , ML models have many different parameters.
# Ex - Features like splitter criteria , Max stab.
# Parameters can be optimized for getting better predictions.
```

```python

from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion": ["squared_error", "friedman_mse", "mae"],
    "splitter": ["best", "random"],
    "max_depth": [None, 10, 20, 30, 40, 50],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model , param_grid= param_grid)
```

```python
grid_tree.fit(X_train , y_train)


# Lets Train our Model
```

```python
grid_tree.best_params_


# Get best parameters for ML Model
```

```python
tree_preds = grid_tree.predict(X_test)

# Get the Performance of the Model
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test , tree_preds)
```
