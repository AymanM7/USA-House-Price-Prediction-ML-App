```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import learning_curve

# Define your feature columns - replace with your actual feature names
features = ['living area', 'number of bedrooms', 'number of bathrooms', 'condition of the house']  # example features

X = df[features]
y = df['Price']

# Generate learning curve data
train_sizes, train_scores, val_scores = learning_curve(
    LinearRegression(), X, y, cv=5, scoring='neg_mean_squared_error',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1, random_state=42
)

# Convert negative MSE to positive MSE by taking the negative
train_errors = -train_scores.mean(axis=1)
val_errors = -val_scores.mean(axis=1)

# Plot the learning curve
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_errors, 'o-', color='blue', label='Training Error (MSE)')
plt.plot(train_sizes, val_errors, 'o-', color='orange', label='Validation Error (MSE)')
plt.xlabel('Training Set Size')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve for Linear Regression Model')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
